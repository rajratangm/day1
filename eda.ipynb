```python
Q1 = df.quantile(0.25)
Q3 = df.quantile(0.75)
IQR = Q3 - Q1
low = Q1 - 1.5 * IQR
high = Q3 + 1.5 * IQR

outliers = (df < low) | (df > high)
outliers.sum().sort_values(ascending=False)   # count per column

df = df[~outliers.any(axis=1)]
```
